```python
max_age = df['age'].max()
max_age_account = df[df['age'] == max_age]
max_age_account
```

```python
min_age = df['age'].min()
min_age_account = df[df['age'] == min_age]
min_age_account
```

```python
box_color = 'red'
whisker_color = 'black' 
plt.boxplot(df['age'], vert=False, patch_artist=True, 
            boxprops=dict(facecolor=box_color, color=whisker_color))
plt.show()
```

```python
plt.violinplot(df['age'], vert=False, showmeans=True, showmedians=True, showextrema=True, widths=0.9, points=100)
plt.title("Age Distribution", fontsize=14)
plt.grid(True)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.show()
```

```python
sns.histplot(data=df, x='age', bins=15, color='hotpink')
plt.xlabel("Age", fontsize=12) 
plt.ylabel("Frequency", fontsize=12) 
plt.title("The Distribution of Age", fontsize=14, fontweight='bold') 
plt.grid(True, linestyle='--', linewidth=0.5, color='gray') 
```

```python
job_categories = df['job'].value_counts()
job_categories
```

```python
job_categories_df = pd.DataFrame({'jobs': job_categories.index, 'counts': job_categories.values})
job_categories_df
```

```python
month_df
```

```python
plt.plot(month_df['months'], month_df['counts'], marker='o', color='green', linestyle='-', linewidth=2, markersize=8)
plt.xlabel('Month', fontsize=12, fontweight='bold', color='blue')
plt.ylabel('Number of respondents', fontsize=12, fontweight='bold', color='blue')
plt.title('Respondents Distribution Over the Months', fontsize=14, fontweight='bold', color='purple')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
